```python
## multilevel inheritence
```

```python
class parents:
    def fun1(self):
        print("This is parent class")
class child(parents):
    def fun2(self):
        print("This is child class")
class grandchild(child):
    def fun3(self):
        print("This is grandchild class")
```

```python
obj=child()
```

```python
obj.fun2()
```

```python
obj.fun1()
```

```python
obj1=grandchild()
```

```python
obj1.fun2()
```

```python
obj1.fun3()
```

```python
obj1.fun1()
```
