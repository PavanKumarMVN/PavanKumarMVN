```python
# Fibonacci number module
def fib(n):
    result=[]
    a,b=0,1
    while b<n:
        result.append(b)
        a,b=b,a+b
    return result
print(fib(1000))
```

## Python Inheritance

```python
class Parent:
    parantAttr=100
    
    def __init__(self):
        print("Calling parent constructor")
        
    def parentMethod(self):
        print("calling parentd method")
        
    def setAttr(self,attr):
        Parent.parentAttr=attr
        
    def getAttr(self):
        print("Parent attribute:",Parent.parentAttr)
        
class Child(Parent):
    def __init(self):
        print("Calling child constructor")
        
    def childMethod(self):
        print("Calling child method")
        
c=Child()
c.childMethod()
```

Overriding Methods

```python
class Parent:
    def myMethod(self):
        print("Calling parents method")
        
class Child(Parent):
    def myMethod(self):
        print("Calling child method")
        
C=Child()
C.myMethod()
```

Overloading Operators

```python
class Vector:
    def __init__(self,a,b):
        self.a=a
        self.b=b
        
    def __str__(self):
        return "Vector (%d,%d)" % (self.a,self.b)
    
    def __add__(self,other):
        return Vector(self.a + other.a,self.b + other.b)
    
v1=Vector(2,6505)
v2=Vector(52,13)
print(v1,v2)
```

Data Hiding

```python
class JustCounter:
    __secretCount=0
    
    def count(self):
        self.__secretCount+=1
        print(self.__secretCount)
    
counter=JustCounter()
counter.count()
counter.count()
print(counter.__secretCount)
```

## Python MultiLevel Inheritance

```python
class Animal:
    def eat(self):
        print("Eating...")
        
class Dog(Animal):
    def bark(self):
        print("Barking...")
        
class BabyDog(Dog):
    def weep(self):
        print("Weeping...")
        
d=BabyDog()
d.eat()
d.bark()
d.weep()
```

## Python Multiple Inheritance

```python
class First(object):
    def __init__(self):
        super(First,self).__init__()
        print("First")
        
class Second(object):
    def __init__(self):
        super(Second,self).__init__()
        print("Second")

class Third(Second,First):
    def __init__(self):
        super(Third,self).__init__()
        print("Third")
        
Third()
```

## Python Exceptions Handling

Handling an Exception

```python
try:
    fh=open("demo","w")
    fh.write("This is my test file for exception Handling!!")
except IOError:
    print("Error: can\'t find file or read data")
else:
    print("Written content in the file sucessfully")
    fh.close()
```

The try-finally Cluse

```python
try:
    fh=open("demo","w")
    fh.write("This is my test file for exception Handling!!")
finally:
    print("Error: can\'t find file or read data")
    fh.close()
```

Argument of an Exception

```python
def temp_convert(var):
    try:
        return int(var)
    except ValueError as Argument:
        print("The argument does not contain numbers\n",Argument)
        
temp_convert("XYZ")
```
