```python
cars=['bmw','kia','lambogini','audi','toyato','ferrari','Bugati']
```

```python
for x in cars:
    print(x)
```

```python
print('these are the cars available in the showroom:\n')
for x in cars:
    print(x)
```

```python
for x in cars:
    if x=='lambogini':
        print(x.upper())
    else:
        print(x.title())
```
