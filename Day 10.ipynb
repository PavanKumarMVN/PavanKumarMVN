## Constructer

```python
__init__  for constructer ----> to construct 
```

```python
class students:
    def __init__(self,name,age,height):
        self.name=name
        self.age=age
        self.height=height
        print("He is in the class")
    def rank(self):
        print(f"{self.name} has passed in the exams")
    def show(self):
        print(f"He is {self.age} and his height is {self.height}")
```

```python
s=students("Varma",26,5.10)
```

```python
s.rank()
```

```python
s.show()
```

```python
s=students("chintu",22,6)
```

```python
s.show()
```
