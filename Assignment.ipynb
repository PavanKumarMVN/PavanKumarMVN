```python
## position argument
```

```python
def make_shirt(size,message):
    print("The size of the T-Shirt is: " + size.upper())
    print("Printed matter on the T-Shirt is: " + message)
    
make_shirt('Large(l)','I love round neck T-shirts')
print()
make_shirt('Medium(m)','In Travelling mood')
```

```python
def make_shirts(size,msg):
    print(f"The T-shirt size is: {size.upper()}")
    print(f"Printed msg is: {msg}")
    
make_shirts("Small(s)","Life gives more opportunities daily \n\t\tSo dont commit suicide")
print()
make_shirts("Extra Large(XL)",'Just do what makes your life happy')
```

```python
## keyword argument
```

```python
def make_shirts(size,msg):
    print(f"The T-shirt size is: {size.upper()}")
    print(f"Printed msg is: {msg}")
    
make_shirts(size="Small(s)",msg="Life gives more opportunities daily \n\t\tSo dont commit suicide")
print()
make_shirts(msg='Just do what makes your life happy',size="Extra large")
```
