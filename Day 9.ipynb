```python
4. Arbitary arguments
```

```python
def greet(*msg):
    for m in msg:
        print(m)
```

```python
greet('hello','how','do','u','do')
```

```python
greet('hello','how','do','u','do','ldjksf','fsdj','wnjf')
```

```python
def printinfo(arg1,*vartuple):
    print("Output is: ")
    print(arg1)
    for var in vartuple:
        print(var)
        
printinfo(215,218,38743,1687)
```

```python
## making a sandwich using function
```

```python
def make_sandwich(item1,item2,item3):
    print("Sandwich is preparing for you")
    print(f"adding {item1} to you sandwich")
    print(f"adding {item2} to you sandwich")
    print(f"adding {item3} to you sandwich")
    print("Your sandwich is prepared for you")
    
make_sandwich('Tomato','cheese','butter and onions')
```
